### New York City Taxi trip Records

Exploración de los datos obtenidos del NYC (Taxi & Limusine commission) [https://www.nyc.gov/site/tlc/about/tlc-trip-record-data.page] 
sobre recorridos de taxis amarillos, verdes y for-hire en Nueva York año 2024

```python
import pandas as pd
import numpy as np
```

### 1. Archivo yellow_tripdata_2024_05.parquet

```python
df1 = pd.read_parquet(r'C:\Users\migue\OneDrive\Escritorio\Datasets Taxi\yellow_tripdata_2024-05.parquet', engine='pyarrow')
df1
```

Observamos 3.7 millones de viajes registrados, con información en 19 columnas que abarca desde tiempo y localización de comienzo y fin del viaje, como distancia, tarifas y cobros extras

```python
df1.isna().sum()
```

Observamos una gran cantidad de nulos representando 404mil viajes donde no se registró cantidad de pasajeros, flag, ni otros cargos. Sin embargo datos importantes como distancia y tarifa están presentes

```python
filas_con_nulos = df1[df1.isnull().any(axis=1)]
filas_con_nulos
```

Observamos 404666 registros en varias columnas nulas, debido a que son información que ingresa el conductor manualmente en el recorrido

```python
duplicated_rows = df1[df1.duplicated()]
duplicated_rows
```

No hay datos duplicados. Se deciden conservar las 19 columnas por su utilidad para el análisis
