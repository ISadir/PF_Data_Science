<h1 align=center> Limpieza de Datos

Preprocesamiento del [original dataset] generando nuevos archivos en la carpeta [data] para su posterior análisis. <br>
La exploración inicial de datos (EDA) y los detalles del análisis de duplicados, faltantes y outliers se encuentra en los archivos EDA1.ipynb y EDA2.ipynb

```python
from dbfread import DBF
import pandas as pd
```

## 1. Zonas de Servicio

importamos librerías y cargamos el primer archivo

```python
df1 = pd.read_csv(r'..\original data\Taxi+_zone_lookup.csv')
df1
```

Limpiamos los registros nulos (detalles en EDA)

```python
df1 = df1.dropna()
df1
```

Cargamos un segundo archivo de zonas para su merge

```python
dbf_file = r'..\original data\taxi_zones.dbf'
table = DBF(dbf_file)
df2 = pd.DataFrame(iter(table))
df2
```

Combinamos ambos dataset en un único archivo por zona

```python
dfzonas = pd.merge(df1, df2, on='LocationID', how='inner')
dfzonas
```

Conservamos sólo las columnas útiles

```python
columnas_quedan = ['LocationID', 'Borough','Zone', 'service_zone', 'Shape_Leng', 'Shape_Area' ]
dfzonas = dfzonas[columnas_quedan]
dfzonas
```

Se guarda el dataframe en un archivo Service Zones.csv

```python
dfzonas.to_csv('..\data\Service Zones.csv', index=False)
```
